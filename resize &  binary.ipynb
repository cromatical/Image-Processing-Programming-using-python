```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np

color_array = plt.imread("/home/lee/Documents//apple.png")
rgb = color_array[:,:,0:3]

h = rgb.shape[0]
w = rgb.shape[1]
print(h, w)

ratio_h = 2
ratio_w = 1.1

new_h = np.int(np.floor(h / ratio_h))
new_w = np.int(np.floor(w / ratio_w))
print(new_h, new_w)

new_rgb = np.zeros((new_h, new_w, 3))

for i in range(new_h):
    for j in range(new_w):
        new_rgb[i,j,0] = rgb[np.int(np.round(ratio_h * i)), np.int(np.round(ratio_w * j)), 0]
        new_rgb[i,j,1] = rgb[np.int(np.round(ratio_h * i)), np.int(np.round(ratio_w * j)), 1]
        new_rgb[i,j,2] = rgb[np.int(np.round(ratio_h * i)), np.int(np.round(ratio_w * j)), 2]

plt.imshow(new_rgb)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np

color_array = plt.imread("/home/lee/Documents//binary_test.png")
gray_img = color_array[:,:,0:3]

h = gray_img.shape[0]
w = gray_img.shape[1]
print(h, w)

ratio_h = 1
ratio_w = 1

new_h = np.int(np.floor(h / ratio_h))
new_w = np.int(np.floor(w / ratio_w))
print(new_h, new_w)

new_gray = np.zeros((new_h, new_w, 3))

for i in range(new_h):
    for j in range(new_w):
        new_gray[i,j,0] = gray_img[np.int(np.round(ratio_h * i)), np.int(np.round(ratio_w * j)), 0]
        new_gray[i,j,1] = gray_img[np.int(np.round(ratio_h * i)), np.int(np.round(ratio_w * j)), 1]
        new_gray[i,j,2] = gray_img[np.int(np.round(ratio_h * i)), np.int(np.round(ratio_w * j)), 2]

new_gray[:,:,1] = new_gray[:,:,0]         
new_gray[:,:,2] = new_gray[:,:,0]         

plt.imshow(new_gray)
```

```python
th = 0.5
new_gray_img = np.where(new_gray <= th , 0.0, 1.0)
plt.imshow(new_gray_img)
```
