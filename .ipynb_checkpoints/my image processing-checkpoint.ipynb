```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img

lenna = img.imread("C:/Users/croma/Desktop///lenna.png")

#   
# def detection(img, mask_size):
#     rows, cols = img.shape
#     detection_h_mask = np.zeros((mask_size,mask_size), np.float32) 
#     detection_h_mask[0][:] = -1
#     detection_h_mask[-1][:] = 1
    
#     detection_v_mask = np.zeros((mask_size,mask_size), np.float32) 
#     detection_v_mask[:][0] = -1
#     detection_v_mask[:][-1] = 1
    
    
#     guard = int((mask_size -1) / 2)
#     detection_img = np.copy(img)
    
#     th = 0.2
    
#     for x in range(guard,rows-guard):
#         for y in range(guard, cols-guard):
#             img_sel = img[x-guard:x-guard+mask_size,y-guard:y-guard+mask_size]
#             v_result = np.sum(detection_v_mask * img_sel)
#             h_result = np.sum(detection_h_mask * img_sel)
#             result = np.sqrt(v_result**2 + h_result**2)
            
#             if result > th:
#                 detection_img[x, y] = 0
#             else:
#                 detection_img[x, y] = 1
        
#     return detection_img

# 
def detection(img, mask_size):
    rows, cols = img.shape
    detection_h_mask = np.zeros((mask_size,mask_size), np.float32) 
    detection_h_mask[0][:] = -1
    detection_h_mask[-1][:] = 1
    
    detection_v_mask = np.zeros((mask_size,mask_size), np.float32) 
    detection_v_mask[:][0] = -1
    detection_v_mask[:][-1] = 1
    
    
    guard = int((mask_size -1) / 2)
    detection_img = np.copy(img)
    
    th = 0.2
    
    for x in range(guard,rows-guard):
        for y in range(guard, cols-guard):
            img_sel = img[x-guard:x-guard+mask_size,y-guard:y-guard+mask_size]
            v_result = np.sum(detection_v_mask * img_sel)
            h_result = np.sum(detection_h_mask * img_sel)
            result = (v_result + h_result) / 2
            
            if result > th:
                detection_img[x, y] = 0
            else:
                detection_img[x, y] = 1
        
    return detection_img

img = lenna[:,:,0]
img1 = detection(3, img)

fig, axes = plt.subplots(1, 2, figsize=(12,12))
axes[0].imshow(img, cmap='gray')
axes[1].imshow(img1, cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img


def erosion(img, mask_size, interations):
    rows, cols = img.shape
    erosion_mask = np.ones((mask_size,mask_size), np.float32) 
    guard = int((mask_size -1) / 2)
    
    for i in range(interations):
        erosion_img = np.copy(img)
        for x in range(guard,rows-guard):
            for y in range(guard, cols-guard):
                erosion_img[x,y] = np.max(img[x-guard:x-guard+mask_size,y-guard:y-guard+mask_size] * erosion_mask)
        img = erosion_img
    return img

def dilation(img, mask_size, interations):
    rows, cols = img.shape
    erosion_mask = np.ones((mask_size,mask_size), np.float32) 
    guard = int((mask_size -1) / 2)
    
    for i in range(interations):
        erosion_img = np.copy(img)
        for x in range(guard,rows-guard):
            for y in range(guard, cols-guard):
                erosion_img[x,y] = np.min(img[x-guard:x-guard+mask_size,y-guard:y-guard+mask_size] * erosion_mask)
        img = erosion_img
    return img

def reduce_image(img, ratio):
    rows = img.shape[0]
    cols = img.shape[1]
    new_h = np.int(np.floor(rows / ratio))
    new_w = np.int(np.floor(cols / ratio))
    
    new_rgb = np.zeros((new_h, new_w, 3))
    for i in range(new_h):
        for j in range(new_w):
            new_rgb[i,j,0] = img[np.int(np.round(ratio * i)), np.int(np.round(ratio * j)), 0]
            new_rgb[i,j,1] = img[np.int(np.round(ratio * i)), np.int(np.round(ratio * j)), 1]
            new_rgb[i,j,2] = img[np.int(np.round(ratio * i)), np.int(np.round(ratio * j)), 2]
    return new_rgb

def image_binary(img, th):
    gray_img = np.where(img <= th , 0.0, 1.0)
    return gray_img

lenna = img.imread("C:/Users/croma/Desktop///.png")
img = reduce_image(lenna, 2)

reduce_img = img[:,:,0]
img1 = erosion(reduce_img, 3, 1)
img2 = image_binary(reduce_img, 0.5)
img3 = erosion(img2, 3, 1)
img4 = erosion(img2, 3, 2)
img5 = dilation(img3, 3, 5)



fig, axes = plt.subplots(1, 4, figsize=(12,12))
# axes[0].imshow(reduce_img, cmap='gray')
# axes[1].imshow(img1, cmap='gray')
axes[0].imshow(img2, cmap='gray')
axes[1].imshow(img3, cmap='gray')
axes[2].imshow(img4, cmap='gray')
axes[3].imshow(img5, cmap='gray')
```

# cell count

```python
def count_cell(img, mask_size):
    rows, cols = img.shape
    count_mask = np.ones((mask_size,mask_size), np.float32) 
    guard = int((mask_size -1) / 2)
    
    count_img = np.copy(img)
    check_img = np.copy(img)
    count = 0
    queue = []

    for x in range(guard,rows-guard):
        for y in range(guard, cols-guard):
            if count_img[x,y] == 1:
                check_img[x, y] == True
            elif check_img[x, y] == True:
                continue
            else:
                queue.append([x,y])
                count += 1
                while queue:
                    idx = queue.pop(0)
                    if check_img[idx[0], idx[1]] == True:
                        continue
                    else:
                        check_img[idx[0], idx[1]] = True
                        for m in range(-1, 2):
                            for n in range(-1, 2):
                                if idx[0]+m == 0 or idx[1]+n == 0:
                                    continue
                                if idx[0]+n == rows or idx[1]+n == cols:
                                    continue
                                if check_img[idx[0]+m, idx[1]+n] == True:
                                    continue
                                elif count_img[idx[0]+m, idx[1]+n] == 1:
                                    check_img[idx[0]+m, idx[1]+n] == True
                                    continue
                                else: 
                                    queue.append([idx[0]+m, idx[1]+n])

    return count
```

```python
def expansion(img, n):
    cols, rows = img.shape
    expansion_mask = np.ones((n,n), np.float32) 
    expansion_img = np.ones((cols*n, rows*n), np.float32)
    # print(expansion_img.shape)
    original_img = np.copy(img)
    
    for x in range(cols):
        for y in range(rows):
            #print(x*n,x*n+2, y*n,y*n+2)
            expansion_img[x*n:x*n+n, y*n:y*n+n] = original_img[x,y] * expansion_mask
    return expansion_img
```

```python
def expand(img, n):
    rows, cols = img.shape
    new_rows = int(rows*n)
    new_cols = int(cols*n)
    
    img = np.copy(img)
    new_img = np.empty([new_rows, new_cols])
    
    idx_new_rows = np.linspace(0, rows-1, new_rows)
    idx_new_rows = idx_new_rows.astype(int)
    idx_new_cols = np.linspace(0, cols-1, new_cols)
    idx_new_cols = idx_new_cols.astype(int)
    
    for x in range(new_rows):
        for y in range(new_cols):
            new_img[x, y] = img[idx_new_rows[x], idx_new_cols[y]]
    
    # idx_new = np.meshgrid(idx_new_rows, idx_new_cols)
    # new_img = img[(idx_new_rows[1], idx_new_cols[0])]
    return new_img
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img

lena = img.imread("C:/Users/croma/Desktop//lena.png")
img = lena[:,:,0]

def interpolation(img,n):
    rows, cols = img.shape
    new_rows = int(rows*n)
    new_cols = int(cols*n)

    copy_img = np.copy(img)
    new_img = np.empty([new_rows, new_cols])

    idx_new_rows = np.linspace(0, rows-1, new_rows)
    # idx_new_rows = idx_new_rows.astype(int)
    idx_new_cols = np.linspace(0, cols-1, new_cols)
    # idx_new_cols = idx_new_cols.astype(int)

    for i, x in enumerate(idx_new_rows[1:-1]):
        i+=1
        for j, y in enumerate(idx_new_cols[1:-1]):
            j+=1
            if len(idx_new_rows) != i or len(idx_new_cols) != j:
                A = copy_img[int(x), int(y)] # LEFT_UP 
                B = copy_img[int(x)+1, int(y)] # RIGHT_UP
                C = copy_img[int(x), int(y)+1] # LEFT_DOWN
                D = copy_img[int(x)+1, int(y)+1] # RIGHT_UP

                E = (1-(x-int(x)))*A + (x-int(x))*B # = A+a(B - A)
                F = (1-(x-int(x)))*C + (x-int(x))*D # = C +a(D -C)
                X = (1- (y-int(y)))*E + (y-int(y))*F # = E + b(F - E)

                new_img[i,j] = X
                
    return new_img

# E = (1-a)*A + a*B # = A+a(B - A)
# F = (1-a)*C + a*D # = C +a(D -C)
# X = (1- b )*E + b*F # = E + b(F - E)

# new_img2 = expand(img, 2)

# fig, axes = plt.subplots(1, 3, figsize=(25,25))
# axes[0].imshow(img, cmap='gray')
# axes[1].imshow(new_img2, cmap='gray')
# axes[2].imshow(new_img, cmap='gray')
```

```python
def subsampling(img, n):
    rows, cols = img.shape
    new_rows = int(rows/n)
    new_cols = int(cols/n)
    
    img = np.copy(img)
    new_img = np.empty([new_rows, new_cols])
    
    idx_new_rows = np.linspace(0, rows-1, new_rows)
    idx_new_rows = idx_new_rows.astype(int)
    idx_new_cols = np.linspace(0, cols-1, new_cols)
    idx_new_cols = idx_new_cols.astype(int)
    
    for x in range(new_rows):
        for y in range(new_cols):
            new_img[x, y] = img[idx_new_rows[x], idx_new_cols[y]]
    
    return new_img
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np

lena = img.imread("C:/Users/croma/Desktop//lenna.png")
array_img = lena[:,:,0]

def warping(img, A, f, direction):
    rows, cols = img.shape
    original_img = np.copy(img)
    img = np.zeros((rows,cols))
    
    for x in range(rows):
        for y in range(cols):
            if direction == 'vertical':
                old_x = A*np.sin(f * (2 * np.pi) * x / rows)
                old_x = old_x.astype(int)
                if (old_x+y < cols) and (old_x+y >= 0):
                     img[x,y] = original_img[x,(old_x+y)]
            
            elif direction == 'horizon':
                old_y = A*np.sin(f * (2 * np.pi) * y / cols)
                old_y = old_y.astype(int)
                if (old_y+x < rows) and (old_y+x >= 0):
                     img[x,y] = original_img[(old_y+x),y]
    return img
```

```python
img = warping(array_img, -100, 2, 'vertical')
plt.imshow(img, cmap='gray')
```

```python
img1 = warping(array_img, 100, 2, 'horizon')
plt.imshow(img1, cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import math

def img_rotation(img, theta):
    rows, cols = img.shape
    angle = math.radians(theta)
    rot_mat = [[math.cos(angle), math.sin(angle)], [-math.sin(angle), math.cos(angle)]]

    angle2 = math.radians(90 - theta)
    h = abs(rows * math.cos(angle)) + abs(cols * math.cos(angle2))
    h = np.round(h).astype(int)  #   

    w = abs(rows * math.cos(angle2)) + abs(cols * math.cos(angle))
    w = np.round(w).astype(int)  #   

    padding_rows = abs((h - rows)) // 2
    padding_cols = abs((w - cols)) // 2
    # print("h :", h, "w :", w)
    # print("rows :", rows, "cols :", cols)
    # print("padding_rows :", padding_rows, "padding_cols :", padding_cols)

    rotation_img = np.zeros((h, w))

    if theta == 90 or theta == 270:
        padding_img = np.zeros((h, h))
        padding_img[padding_rows:rows + padding_rows, :] = img[:, :]
        mid_x = math.ceil((h + 1) / 2)
        mid_y = math.ceil((h + 1) / 2)
    else:
        padding_img = np.zeros((h, w))
        padding_img[padding_rows:rows + padding_rows, padding_cols:cols + padding_cols] = img[:, :]
        mid_x = math.ceil((w + 1) / 2)
        mid_y = math.ceil((h + 1) / 2)

    # print("padding.shape :", padding_img.shape)

    for x in range(h):
        for y in range(w):
            new_xy = np.array([y, h - 1 - x])
            new_xy_shifted = new_xy - np.array([mid_x, mid_y])

            old_xy_shifted = np.matmul(rot_mat, new_xy_shifted)
            old_xy = old_xy_shifted + np.array([mid_x, mid_y])

            old_pxl_r = h - 1 - old_xy[1]
            old_pxl_c = old_xy[0]

            old_pxl_r = np.round(old_pxl_r).astype(int)
            old_pxl_c = np.round(old_pxl_c).astype(int)

            if theta == 90:
                if (old_pxl_r < h) and (old_pxl_c < h):
                    rotation_img[x, y] = padding_img[old_pxl_r + 25, old_pxl_c]
            elif theta == 270:
                if (old_pxl_r < h) and (old_pxl_c < h):
                    rotation_img[x, y] = padding_img[old_pxl_r - 25, old_pxl_c]
            else:
                if (old_pxl_r >= 0 and old_pxl_r < h) and (old_pxl_c >= 0 and old_pxl_c < w):
                    rotation_img[x, y] = padding_img[old_pxl_r, old_pxl_c]
    return rotation_img


def img_warping(img, amplitude, frequency):
    o_img = np.copy(img)
    rows, cols = img.shape
    warping_img = np.zeros((rows, cols))

    A = amplitude * 2
    f = frequency

    i = np.linspace(0, cols - 1, cols)
    new_rows = np.linspace(A * -np.sin(f * (2 * np.pi) * i / cols),
                           A * np.sin(f * (2 * np.pi) * i / cols),
                           rows)
    new_rows = new_rows.astype(int)

    for x in range(rows):
        for y in range(cols):
            if x + new_rows.T[y][x] < rows and x + new_rows.T[y][x] >= 0:
                warping_img[x, y] = o_img[x + new_rows.T[y][x], y]

    return warping_img


seven = img.imread("C:/Users/croma/Desktop//seven.jpg")
img1 = seven[:, :, 0]

lenna = img.imread("C:/Users/croma/Desktop//lenna.png")
img2 = lenna[:,:,0]

result1 = img_rotation(img1, 150)
result2 = img_warping(img2, 150, 2)
```
